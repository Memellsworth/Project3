```python
csv_names = []
for airline in airlines:
    try:

#filter airport ot houston 
    #click airport dropdown
        time.sleep(10)
        driver.find_element("id","cboAirline").click()
        #clickdropdown an then find airline name 
        driver.find_element("xpath", "//*[contains(text(),'" + airline + "')]").click()

        #click submit button
        driver.find_element("id", "btnSubmit").click()
        #wait for page to load
        time.sleep(5)
        
        # #export the csv
        driver.find_element("id", "DL_CSV").click()
        #wait for it to download 
        time.sleep(20)
        #CHANGE FOR YOUR DOWNLOAD PATH 
        download_dir = r"C:\..\..\Downloads"
        file_name = "Detailed_Statistics_Departures.csv"
        file_path = os.path.join(download_dir, file_name)
    
        # Set the destination directory CHANGE TO YOUR FILE PATH 
        destination_dir = r"C:\Users\memel\OneDrive\Desktop\Project3\CSV"
        os.makedirs(destination_dir, exist_ok=True)

        # Rename and move the downloaded file to the destination directory
        destination_file = os.path.join(destination_dir, airline + ".csv")
        shutil.move(file_path, destination_file)
        csv_names.append(f"{airline}.csv")
        time.sleep(5)
    except Exception as e:
        print(f"Error occured for airline {airline} skipping.. Error message {str(e)}")
```
